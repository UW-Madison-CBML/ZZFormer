```python
#!/usr/bin/env python3
"""
Convert FASTA files to CSV-style text files.

Input  (FASTA):
    >LTR_Copia_0
    tgcatatcccgcgtttag

Output (TXT):
    LTR_Copia,TGCATATCCCGCGTTTAG

The trailing "_<index>" in the header is stripped.
Multi-line sequences are joined into a single line.
"""

ordersss={'Structural_RNA', 'SINE', 'DNA', 'LTR', 'LINE', 'Other', 'Satellite', 'PLE', 'RC'}
repeated_label={"Structural_RNA_Structural_RNA": "Structural_RNA",
"SINE_SINE": "SINE",
"DNA_DNA": "DNA",
"LTR_LTR": "LTR",
"LINE_LINE": "LINE",
"Other_Other": "Other",
"Satellite_Satellite": "Satellite",
"PLE_PLE": "PLE",
"RC_RC": "RC",}


import os
import re
import sys
from glob import glob

# Matches a trailing "_<digits>" at the end of the header, e.g. "_0", "_123"
TRAIL_IDX = re.compile(r'_\d+$')

def convert_fasta(in_path, out_path):
    with open(in_path, 'r') as fin, open(out_path, 'w') as fout:
        label = None
        seq_parts = []

        def flush():
            if label is not None and seq_parts:
                seq = ''.join(seq_parts).upper()
                
                fout.write(f"{label},{seq}\n")

        for line in fin:
            line = line.strip()
            if not line:
                continue
            if line.startswith('>'):
                # write previous record
                flush()
                header = line[1:].split()[0]            # drop ">" and any description
                label = TRAIL_IDX.sub('', header)       # drop trailing _<int>
                if label in repeated_label:
                    label = repeated_label[label]
                seq_parts = []
            else:
                seq_parts.append(line)
        flush()  # last record

def main(roots):
    # Default: process the three subfolders in the current directory
    if not roots:
        roots = ['mntedb', 'repbase', 'repetdb']
    infolder_name= "/Users/kkumari/Desktop/TDL_DMS_Data/TERRIER_Labelling_system/new_splits_FromLevi/DeepTE"
    for root in roots:
        if not os.path.isdir(f"{infolder_name}/{root}"):
            print(f"Skipping (not a dir): {root}")
            continue
        for fa in sorted(glob(os.path.join(f"{infolder_name}/{root}", '*.fasta'))):
            txt = fa.replace("new_splits_FromLevi/DeepTE/","DeepTE/").replace('.fasta', '.txt')      # replace .fasta with .txt
            print(f"{fa}  ->  {txt}")
            convert_fasta(fa, txt)

```

```python
if __name__ == '__main__':
    main([])
```
